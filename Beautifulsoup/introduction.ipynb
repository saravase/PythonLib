```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

url="https://www.pythonforbeginners.com"
content =urlopen(url).read()
soup=BeautifulSoup(content)
print(soup.prettify)
```

```python
links=soup.find_all('a')
links
```

```python
for link in links:
    print(link.get("href"))
```
